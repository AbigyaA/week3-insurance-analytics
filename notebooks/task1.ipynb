```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')
```

```python
# Load the dataset (adjust delimiter if needed)
df = pd.read_csv('MachineLearningRating_v3.txt', delimiter='|') 
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_") 
df.head()
```

```python
with open('MachineLearningRating_v3.txt') as f:
    for _ in range(5):
        print(f.readline())
```

```python
# Show data types and summary
print(df.info())
print(df.describe(include='all'))

# Check missing values
print(df.isnull().sum().sort_values(ascending=False))
```

```python
# Distribution of Total Claims
sns.histplot(df['totalclaims'].dropna(), kde=True)
plt.title("Distribution of Total Claims")
plt.show()

# Distribution of Total Premium
sns.histplot(df['totalpremium'].dropna(), kde=True)
plt.title("Distribution of Total Premium")
plt.show()

# Distribution of Custom Value Estimate
sns.histplot(df['customvalueestimate'].dropna(), kde=True)
plt.title("Custom Value Estimate Distribution")
plt.show()

# Bar plot: Gender
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()

# Top 10 Vehicle Types
sns.countplot(y='vehicletype', data=df, order=df['vehicletype'].value_counts().head(10).index)
plt.title("Top 10 Vehicle Types")
plt.show()
```

```python
# Avoid division by zero
df['loss_ratio'] = df.apply(
    lambda row: row['totalclaims'] / row['totalpremium'] if row['totalpremium'] > 0 else 0,
    axis=1
)
```

```python
# Boxplot: Loss Ratio by Gender
sns.boxplot(x='gender', y='loss_ratio', data=df)
plt.title("Loss Ratio by Gender")
plt.show()

# Average Loss Ratio by Province
province_loss = df.groupby('province')['loss_ratio'].mean().sort_values(ascending=False)
province_loss.plot(kind='bar', figsize=(10, 5), title='Average Loss Ratio by Province')
plt.ylabel("Loss Ratio")
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df[['totalpremium', 'totalclaims', 'customvalueestimate', 'loss_ratio']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Scatter plot: Premium vs Claims
sns.scatterplot(x='totalpremium', y='totalclaims', hue='province', data=df)
plt.title("Premium vs Claims by Province")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Custom Value Estimate
sns.boxplot(y='customvalueestimate', data=df)
plt.title("Outlier Detection: Custom Value Estimate")
plt.show()

# Total Claims
sns.boxplot(y='totalclaims', data=df)
plt.title("Outlier Detection: Total Claims")
plt.show()
```

```python
monthly_trend = df.groupby('transactionmonth')[['totalclaims', 'totalpremium']].sum()
monthly_trend.plot(figsize=(12,6), title='Monthly Total Claims and Premiums')
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
make_claims = df.groupby('make')['totalclaims'].mean().sort_values(ascending=False).head(10)
make_claims.plot(kind='barh', figsize=(10,6), title='Top 10 Makes by Average Claims')
plt.xlabel("Average Total Claims")
plt.show()
```

```python
pivot = df.pivot_table(index='vehicletype', columns='gender', values='totalclaims', aggfunc='mean')
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Average Claims by Vehicle Type and Gender")
plt.show()
```
